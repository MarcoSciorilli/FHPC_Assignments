```python
import matplotlib.pyplot as plt 
import numpy as np
```

```python
data_all = np.loadtxt('/home/marco/Desktop/Assignment_2/Assignment_2_all.txt') 
data_first = np.loadtxt('/home/marco/Desktop/Assignment_2/Assignment_2_first.txt')
```

```python
data_temp=np.zeros((20,2))
for i in range(20):
    for n in range(int(len(data_first)/20)):
        data_first[i,1] = data_first[i,1] + data_first[i+n*20,1]
    #data_all[i,1] = data_all[i,1]/(len(data_all)/20)
    data_temp[i,1] = data_first[i,1]/(int(len(data_first)/20+1))
    data_temp[i,0] = data_first[i,0]
data_first=data_temp 
data_temp=np.zeros((20,2))
```

```python
data_temp=np.zeros((20,2))
for i in range(20):
    for n in range(int(len(data_all)/20)):
        data_all[i,1] = data_all[i,1] + data_all[i+n*20,1]
    data_temp[i,1] = data_all[i,1]/(int(len(data_all)/20+1))
    data_temp[i,0] = data_all[i,0]
data_all=data_temp 
data_temp=np.zeros((20,2))
```

```python
data_all = np.loadtxt('/home/marco/Desktop/Assignment_3/strongopenMP.txt') 
data_first = np.loadtxt('/home/marco/Desktop/Assignment_3/strongMPI.txt')
```

```python
axes = plt.gca()
axes.set_xlim([1,20])
axes.set_ylim([0,14])
plt.plot(data_all[:,0],data_all[:,1], label="OpenMp", linewidth=1)
plt.plot(data_first[:,0],data_first[:,1], label="MPI",linewidth=1)
plt.legend()
plt.grid()
plt.xlabel('# of processors')
plt.ylabel('Time')
plt.savefig('Time.svg', format='svg', dpi=1200)
plt.show()
```

```python
axes = plt.gca()
axes.set_xlim([1,20])
axes.set_ylim([1,17.5])
plt.plot(data_all[:,0],data_all[0,2]/data_all[:,2], label="OpenMP", linewidth=1)
plt.plot(data_first[:,0],data_first[0,2]/data_first[:,2], label="MPI",linewidth=1)
plt.legend()
plt.grid()
plt.xlabel('# of processors')
plt.ylabel('Speedup')
plt.savefig('Speedup.svg', format='svg', dpi=1200)
plt.show()
```

```python
axes = plt.gca()
axes.set_xlim([1,20])
axes.set_ylim([1,12])
plt.plot(data_all[:,0],data_all[0,2]/data_all[:,2], label="OpenMP", linewidth=1)
plt.plot(data_first[:,0],data_first[0,2]/data_first[:,2], label="MPI",linewidth=1)
plt.plot(data_all[:,0],data_all[:,0], label="Ideal",linewidth=1)
plt.legend()
plt.grid()
plt.xlabel('# of processors')
plt.ylabel('Speedup')
plt.savefig('Speedup.svg', format='svg', dpi=1200)
plt.show()
```

```python
axes = plt.gca()
axes.set_xlim([1,20])
axes.set_ylim([0,7])
plt.plot(data_all[:,0],data_all[:,1]-data_all[0,1]/data_all[:,0], label="OpenMP", linewidth=1)
plt.plot(data_first[:,0],data_first[:,1]-data_first[0,1]/data_first[:,0], label="MPI",linewidth=1)
plt.legend()
plt.grid()
plt.xlabel('# of processors')
plt.ylabel('Overhead')
plt.savefig('Overhead.svg', format='svg', dpi=1200)
plt.show()
```

```python
speedup_first=np.zeros((19,1))
speedup_all=np.zeros((20,1))
for i in range(19):
    speedup_first[i]=data_first[0,1]/data_first[i,1]
for i in range(20):
    speedup_all[i]=data_all[0,1]/data_all[i,1]
serialf_first=np.zeros((19,1))
serialf_all=np.zeros((20,1))
for i in range(19):
    serialf_first[i]=((1/speedup_first[i])-1/(i+1))/(1-1/(i+1))
for i in range(20):
    serialf_all[i]=((1/speedup_all[i])-1/(i+1))/(1-1/(i+1))
```

```python
axes = plt.gca()
axes.set_xlim([2,20])
axes.set_ylim([0,0.1])
plt.plot(data_all[:,0],serialf_all[:], label="OpenMP", linewidth=1)
plt.plot(data_first[:,0],serialf_first[:], label="MPI",linewidth=1)
plt.legend()
plt.grid()
plt.xlabel('# of processors')
plt.ylabel('serial fraction')
plt.savefig('Serial_fraction.svg', format='svg', dpi=1200)
plt.show()
```
